```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models



data = r'C:\Users\SYAMNARAYANAN\OneDrive\Desktop\visual code\Dataset of Tuberculosis Chest X-rays Images'

dataprocess = ImageDataGenerator(
    rescale=1/255,
    validation_split=0.2  
)
```

```python
image_size = (256, 256)
batch_size = 32



train_generator = dataprocess.flow_from_directory(
    data,
    target_size=image_size,
    class_mode='binary',
    subset='training'
)

```

```python
val_generator = dataprocess.flow_from_directory(
    data,
    target_size=image_size,
    batch_size=batch_size,
    class_mode='binary',
    subset='validation',
    shuffle=False)
```

```python
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(256, 256, 3)),
    layers.MaxPooling2D((2, 2)),
    
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    
    layers.Flatten(),
    layers.Dense(512, activation='relu'),
    
    layers.Dropout(0.2),
    layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python

# early_stopping = EarlyStopping(
#     monitor='val_loss',
#     patience=5,
#     restore_best_weights=True
# )
```

```python

history = model.fit(
    train_generator,
    epochs=15,
    validation_data=val_generator
)

loss, accuracy = model.evaluate(val_generator)
print(f"Validation Accuracy: {accuracy*100:.2f}%")
```

```python

model.save('tb_model.h5')
print("Model saved as tb_model.h5")
```
